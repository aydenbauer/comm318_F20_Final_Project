In this notebook I will include data about Netflix and its seasonal trends pattern.

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
NetflixTimeline_df = pd.read_csv('../data/NetflixTimeline.csv', skiprows=1)
```

```python
NetflixTimeline_df.columns = ["date","value"]
```

```python
NetflixTimeline_df = NetflixTimeline_df.sort_values("value", ascending = False) 
```

```python
NetflixTimeline_df=pd.read_csv('../data/NetflixTimeline.csv', skiprows=1, index_col=0, parse_dates=True)
```

```python
NetflixTimeline_df.plot(figsize=(12,4))
```

This is the time-line for Netflix searches throughout the 2020 year. The peak takes place during the third week of March and the overall values are higher after March than the other previous months. 

Before making any conclusions I want to compare this information against information from previous years. 

```python
NetflixTimeline2019_df = pd.read_csv('../data/NetflixTimeline2019.csv', skiprows=1)
```

```python
NetflixTimeline2019_df.columns = ["date","value"]
```

```python
NetflixTimeline2019_df = NetflixTimeline2019_df.sort_values("value", ascending = False) 
```

```python
NetflixTimeline2019_df.head()
```

```python
NetflixTimeline2019_df=pd.read_csv('../data/NetflixTimeline2019.csv', skiprows=1, index_col=0, parse_dates=True)
```

```python
NetflixTimeline2019_df.plot(figsize=(12,4))
```

This is the time-line from 2019 of Netflix searches. I wanted to compare the two years to see if they were different and if COVID had an impact or if they were the same and the peaks were comparable. In this chart we can assume the peak is because of the holidays and I notice that there is no peak during early spring around March and April. Therefore the two years have different trends.

```python
xmax2019=NetflixTimeline2019_df['2019'].idxmax()
xmax2020=NetflixTimeline_df['2020'].idxmax()

rcParams['figure.figsize'] = 12 , 4
plt.plot(NetflixTimeline_df, label='2019', color='purple')
plt.plot(NetflixTimeline2019_df, label='2018')
plt.legend ()

plt.xlabel('Year')
plt.ylabel('Popularity')
plt.title('Netflix Search Trend')
         

plt.vlines(xmax2019, 0, 110, linestyle='dashed', color='gray')
plt.text(xmax2019, 100, '2019 Peak 12/23/18')

plt.vlines(xmax2020, 0, 110, linestyle='dashed', color='gray')
plt.text(xmax2020, 100, '2020 Peak 3/22/20')
```

In a more easy to read graph like the one above we can see that the two years brought different trends. 2019-2020 saw the holiday peak however it was not as large as the peak that took place in March.
