I will do further research to see what was happening with Netflix during this week - if a new show had recently come out or if people just started staying home therefore they were in general watching more television and movies on Netflix. Additionally, the peak took place during the holiday season which is not surprising because in past years this is when the peak occurs due to people spending time together and it being a popular time for new shows and movies to air.
